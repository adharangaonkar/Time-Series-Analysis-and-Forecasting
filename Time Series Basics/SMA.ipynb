### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing Data

```python
close = pd.read_csv('../data/sp500_close.csv')
```

```python
close.head()
```

```python
goog = close[['GOOG']].dropna()
```

```python
goog
```

```python
goog.plot()
```

```python
plt.hist(goog, bins = 20)
```

### SMA - 10

```python
goog['SMA-10'] = goog['GOOG'].rolling(10).mean()
```

```python
goog.plot(figsize =(15,9))
```

### SMA - 50

```python
goog['SMA-50'] = goog['GOOG'].rolling(50).mean()
```

```python
goog.plot(figsize =(15,9))
```
