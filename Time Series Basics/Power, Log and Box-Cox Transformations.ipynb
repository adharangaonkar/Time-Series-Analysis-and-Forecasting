```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy.stats import boxcox
```

```python
data = pd.read_csv("../data/AirPassengers.csv", index_col = 'Month', parse_dates = True)
```

```python
data.head()
```

```python
plt.figure(figsize = (12, 8))
data['#Passengers'].plot()
```

### Square Root

```python
data['Sqrt#Passengers'] = np.sqrt(data['#Passengers'])
```

```python
plt.figure(figsize = (12, 8))
data['Sqrt#Passengers'].plot()
```

The data appears to be squashed, but not as much as needed

### Log transform

```python
data['Log#Passengers'] = np.log(data['#Passengers'])
```

```python
plt.figure(figsize = (12, 8))
data['Log#Passengers'].plot()
```

### Box Cox Transform

```python
boxcox_data, lam = boxcox(data['#Passengers'])
```

Here **lam** is the **Lambda** value for the transformation

```python
data['BoxCox#Passengers'] = boxcox_data
```

```python
plt.figure(figsize = (12, 8))
data['BoxCox#Passengers'].plot()
```

The data is squashed. More so than the square root transform, but lesser than the Log transform
